# Importing Libraries & Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor, ExtraTreesRegressor, BaggingRegressor, AdaBoostRegressor
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS as stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/final_dataset.csv')
df.head()
```

```python
features = df[['minplayers', 'maxplayers', 'minage', 'languagedependence', 'minplaytime', 'maxplaytime',
               'boardgamecategory_cnt', 'boardgamemechanic_cnt']]
poly = PolynomialFeatures(degree = 2)
output_array = poly.fit_transform(features)
output_df = pd.DataFrame(output_array, columns = poly.get_feature_names(features.columns))
output_df.drop(columns = ['1', 'minplayers', 'maxplayers', 'minage', 'languagedependence', 
               'boardgamecategory_cnt', 'boardgamemechanic_cnt'], inplace = True)
df = pd.concat([df, output_df], axis=1, sort=False)
```

```python
cvec = CountVectorizer(stop_words = stopwords, token_pattern=r'\b[^\d\W]+\b')
fit_cvec = cvec.fit_transform(df['description'])
cvec_df = pd.DataFrame(fit_cvec.toarray(), columns = cvec.get_feature_names())
df = pd.concat([df, cvec_df], axis=1, sort=False)
```

```python
X = df._get_numeric_data().drop(columns = 'avgweight')
y = df['avgweight']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 827)
```

# Modeling

## Linear Regression Model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_preds = lr.predict(X)
mean_squared_error(y, lr_preds)
```

```python
cross_val_score(lr, X_train, y_train, cv = 5).mean()
```

```python
lr.score(X_test, y_test)
```

```python
def mse(model):
    print(f'Train RMSE: {np.sqrt(mean_squared_error(y_train, model.predict(X_train)))}')
    print(f'Test RMSE: {np.sqrt(mean_squared_error(y_test, model.predict(X_test)))}')
    print(f'Overall RMSE: {np.sqrt(mean_squared_error(y, model.predict(X)))}')

mse(lr)
```

```python
def best_fit_line(model):
    plt.figure(figsize=(12,9))

    plt.scatter(model.predict(X), y, s=5, color='blue', alpha = 0.5)

    plt.plot([0, np.max(y)],
             [0, np.max(y)],
             color = 'black')

    plt.xlabel("Predicted Values: $\hat{y}$", fontsize = 20)
    plt.ylabel("Actual Values: $y$", fontsize = 20)
    plt.title('Predicted Test Values vs. Actual Test Values', fontsize = 24);
best_fit_line(lr)
```

```python
def resid_plot(model):
    plt.figure(figsize=(12,9))

    resids = y_test - model.predict(X_test)
    sns.scatterplot(x = range(len(resids)),y = resids, alpha = .5)
    plt.axhline(0, c = "r", alpha = .7);
resid_plot(lr)
```

## LASSO Model

```python
pipe = Pipeline(steps = [
    ('sc', StandardScaler()),
    ('lasso', LassoCV())
])

params = {
    'lasso__alphas' : [(np.logspace(-2, 0, 100)), (np.logspace(-1, 1, 200))],
    'lasso__n_alphas' : [100, 150, 200, 250],
    'lasso__max_iter' : [1000, 2000]
 }

gs_lasso = GridSearchCV(pipe, params, cv = 5)
gs_lasso.fit(X_train, y_train)
gs_lasso.best_estimator_
```

```python
cross_val_score(gs_lasso.best_estimator_, X_train, y_train, cv = 5).mean()
```

```python
gs_lasso.score(X_test, y_test)
```

```python
mse(gs_lasso)
```

```python
best_fit_line(gs_lasso)
```

```python
resid_plot(gs_lasso)
```
