```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
# importing dataset of the top 20,000 games on Board Game Geek from:
# https://www.kaggle.com/extralime/20000-boardgames-dataset
df = pd.read_csv('../data/boardgames1.csv')
df.head()
```

```python
df.info()
```

There are 52 columns in this dataset, many of which will be dropped shortly as they aren't relevant to the problem statement. Only two of the columns have any null data: "min_community" and "max_community." Those columns likely won't be included in the final model, as it's not data a designed would have at hand when designing a game.

```python
df.keys()
```

```python
keep_cols = [
    'objectid', 'name', 'minplayers', 'maxplayers', 'minplaytime',
    'maxplaytime', 'minage', 'languagedependence', 'avgweight', 'label',
    'boardgamecategory_cnt', 'boardgamemechanic_cnt', 'boardgamecategory',
    'boardgamemechanic', 'description', 'gamelink'
]
df = df[keep_cols]
df.head()
```

```python
sns.heatmap(df.corr()[['avgweight']].sort_values('avgweight'), cmap = 'coolwarm', annot = True)
```

```python
df['boardgamemechanic'] = df[['boardgamemechanic']].applymap(lambda x: x.replace("'", ''))
df['boardgamemechanic'] = df[['boardgamemechanic']].applymap(lambda x: x.replace("[", ''))
df['boardgamemechanic'] = df[['boardgamemechanic']].applymap(lambda x: x.replace("]", ''))
df['boardgamemechanic']
```

```python
df['boardgamecategory'] = df[['boardgamecategory']].applymap(lambda x: x.replace("'", ''))
df['boardgamecategory'] = df[['boardgamecategory']].applymap(lambda x: x.replace("[", ''))
df['boardgamecategory'] = df[['boardgamecategory']].applymap(lambda x: x.replace("]", ''))
df['boardgamecategory']
```

```python
mechanics = df['boardgamemechanic'].str.get_dummies(sep=',')
```

```python
mechanics
```

```python
base_url = 'https://boardgamegeek.com/browse/boardgamemechanic'
res = requests.get(base_url)
soup = BeautifulSoup(res.content)
```

```python
soup
```

```python
mechanic = soup.find_all('a')[1:]
```

```python
mechanics_list = []
for line in mechanic:
    observation = line.text
    mechanics_list.append(observation)
mechanics_list
```

```python
mechanics_list = [i for i in mechanics_list if i] 
```

```python
mechanics
```

```python
features_list = []
for key in mechanics.keys():
    key.strip()
    if key in mechanics_list:
        features_list.append(key)
```

```python
len(features_list)
```

```python
len(mechanics_list)
```

```python
mechanics = mechanics[features_list]
```

```python
mechanics
```

```python
df = pd.concat([df, mechanics], axis=1, join='inner')
df.shape
```
